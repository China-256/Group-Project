Brief description of the variables in this dataset.

1. **Track**: Name or title of the song.  
2. **Album Name**: Name of the album in which the track is featured.  
3. **Artist**: Name of the performer(s) or band that recorded the track.  
4. **Release Date**: When the track or album was officially released.  
5. **ISRC (International Standard Recording Code)**: A unique identifier for the track, used to track royalties and usage.  
6. **All Time Rank**: The ranking of the track based on cumulative performance metrics in Spotify.  
7. **Track Score**: An aggregate score that represents the overall performance or popularity of the track.  
8. **Spotify Streams**: Total number of times the track has been streamed on Spotify.  
9. **Spotify Playlist Count**: The number of Spotify playlists that include this track.  
10. **Spotify Playlist Reach**: The estimated total audience or potential listeners of the playlists where the track is featured.  
11. **Spotify Popularity**: A metric provided by Spotify, ranging from 0 to 100, reflecting the current popularity of the track.  
12. **YouTube Views**: Total number of views the track’s associated video has received on YouTube.  
13. **YouTube Likes**: Number of likes the track’s video has received on YouTube.  
14. **TikTok Posts**: Number of TikTok videos that use the track as background music.  
15. **TikTok Likes**: Total likes received by all TikTok posts featuring the track.  
16. **TikTok Views**: Total views of TikTok posts featuring the track.  
17. **YouTube Playlist Reach**: The total potential audience of YouTube playlists that include the track.  
18. **Apple Music Playlist Count**: The number of Apple Music playlists that feature the track.  
19. **AirPlay Spins**: Number of times the track was played via AirPlay (radio airplay or streaming service broadcasts).  
20. **SiriusXM Spins**: Number of times the track has been played on SiriusXM satellite radio.  
21. **Deezer Playlist Count**: The number of playlists on Deezer that include the track.  
22. **Deezer Playlist Reach**: Estimated total audience or reach of the playlists on Deezer that feature the track.  
23. **Amazon Playlist Count**: Number of playlists on Amazon Music that include the track.  
24. **Pandora Streams**: Total streams of the track on Pandora.  
25. **Pandora Track Stations**: The number of Pandora stations (custom or curated) that include the track.  
26. **Soundcloud Streams**: Total number of streams the track has received on SoundCloud.  
27. **Shazam Counts**: Number of times the track has been identified using Shazam.  
28. **TIDAL Popularity**: A measure of the track's popularity on TIDAL, usually reflecting plays, likes, and shares.  
29. **Explicit Track**: A boolean value indicating whether the track contains explicit content (e.g., profanity or adult themes).

# Analysis

## What is the overall distribution of ratings across different songs?

```python
sns.histplot(data = data_cleaned, x='Spotify Popularity', kde=True, bins=20, color='blue')
plt.title("Distribution of Spotify Popularity", fontsize=10)
plt.xlabel('Spotify Popularity')
plt.ylabel('Frequency')
plt.show()
```

The histogram of Spotify Popularity and its corresponding frequency presents the following observations:

**Central Tendency:**

The distribution peaks around a Spotify Popularity score of 60–70, indicating that most tracks fall within this popularity range.

**Shape of Distribution:**

The graph shows a roughly normal distribution, although it is slightly skewed to the left. This suggests that while most tracks are moderately popular, there are fewer highly unpopular tracks.

### Identify outliers

```python
Q1 = data_cleaned['Spotify Popularity'].quantile(0.25)  # First quartile (25th percentile)
Q3 = data_cleaned['Spotify Popularity'].quantile(0.75)  # Third quartile (75th percentile)
IQR = Q3 - Q1                                # Interquartile range
lower_bound = Q1 - 1.5 * IQR                 # Lower bound
upper_bound = Q3 + 1.5 * IQR                 # Upper bound

outliers = data_cleaned[(data_cleaned['Spotify Popularity'] < lower_bound) | (data_cleaned['Spotify Popularity'] > upper_bound)]
```

```python
print(f'The lower bound is {lower_bound} and the upper bound is {upper_bound} in the dataset')
```

Tracks with popularity scores below 45.5 may indicate niche tracks or tracks with limited reach, while scores above 89.5 suggest that they are viral content.

```python
outliers_sorted = outliers.sort_values(by='Spotify Streams', ascending =False)
print(outliers_sorted[['Track', 'Artist', 'Spotify Streams', 'Spotify Popularity']])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_cleaned['Spotify Popularity'], data_cleaned['Spotify Streams'], alpha=0.5)
plt.title("Relationship between Popularity and Streams")
plt.xlabel("Spotify Popularity")
plt.ylabel("Spotify Streams in billions")
plt.grid(True)
plt.show()

# Correlation coefficient
corr_pop_streams = data_cleaned[['Spotify Popularity', 'Spotify Streams']].corr().iloc[0, 1]
print(f"Correlation between Popularity and Streams: {corr_pop_streams}")
```

Having a correlation of 0.44 means that it is a moderate correlation, meaning that the more streamed a song, the more popular it is. However, there are other factors that can influence popularity rather than only the number of streams (like social media presence or marketing). to go deeper in this topic we will study the user engagement.

## Top Performers in 2024

we want to know the most played performers in 2024, which have also released albums or singles in 2024

```python
data_2024 = data_cleaned[data_cleaned['Release Date'].dt.year == 2024]

artist_streams_2024 = data_2024.groupby('Artist')['Spotify Streams'].sum().reset_index()

top_2024_artists = artist_streams_2024.sort_values(by='Spotify Streams', ascending=False).head(10)

top_2024_artists
```

In this table, we can see the most streamed artists that released at least a single in 2024. However, the Spotify Streams is an acummulative variable, meaning that if a single or album was released at the beginning of the year, it will accumulate more streams during the year than an album that was released later. We proceed to study the release date of the albums.

```python
data_cleaned['Year'] = data_cleaned['Release Date'].dt.year
data_cleaned['Month'] = data_cleaned['Release Date'].dt.month

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

data_cleaned['Season'] = data_cleaned['Month'].apply(get_season)
```

```python
season_counts = data_cleaned['Season'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(season_counts, labels=season_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'salmon', 'gold'])
plt.title('Distribution of Song Releases by Season')
plt.show()
```

As we can see the most popular season according to historical data is Spring, followed by Summer and Winter. However, the results are very close to each other, which could mean that the season of the release is not that relevant.

```python
songs_per_year = data_cleaned.groupby('Year').size().reset_index(name='Number of Songs')

# Plotting the number of songs released per year
plt.figure(figsize=(10, 6))
plt.plot(songs_per_year['Year'], songs_per_year['Number of Songs'], marker='o', linestyle='-', color='b')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Number of Songs Released')
plt.title('Number of Songs Released Per Year')

# Display the plot
plt.tight_layout()
plt.grid(True)
plt.show()
```

as we can see in this graph we have an increase trend, however for some reason, in 2024 we have a massive sink. This awake some curiosity and we decided to study a little bit further what could have happened. The initial approach was to think that 2024 was not a year with many music releases. So we study the 2024 release date.

```python
year = 2024
data_year = data_cleaned[data_cleaned['Year'] == year]

# Find the latest (maximum) month for that year
latest_month = data_year['Month'].max()

print(f"The latest month in {year} was June")
```

This means that the results obtained from our dataset are going to be only for the first semester of the year. This have some major implications:
- Songs released in January had the opportunity to gather more streams that songs released in May or June.
- Other songs or albums released from July to December are not considered in this study.

```python
year_to_analyze = 2024

# Filter the data for the selected year
data_year = data_cleaned[data_cleaned['Year'] == year_to_analyze]

# Count the number of songs released per month
songs_per_month = data_year.groupby('Month').size().reset_index(name='Number of Songs')

# Sort by month to maintain correct order
songs_per_month = songs_per_month.sort_values(by='Month')

# Map month numbers to month names for better readability
import calendar
songs_per_month['Month'] = songs_per_month['Month'].apply(lambda x: calendar.month_name[x])

# Display the result
print(songs_per_month)

# Plot the number of songs released per month
plt.figure(figsize=(10, 5))
plt.bar(songs_per_month['Month'], songs_per_month['Number of Songs'], color='blue')

# Customize labels and title
plt.xlabel('Month')
plt.ylabel('Number of Songs Released')
plt.title(f'Number of Songs Released Per Month in {year_to_analyze}')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```

The month with more releases is April, followed by March and February (this correspond to the beginning of Spring as the pie chart before). However, it seems that this dataset was created in June, which explains the missing data of the rest of the year.

The Spotify popularity score is a metric that quantifies the popularity of a song or artist on Spotify. This score is calculated based on various factors, including the number of streams, user interactions like saves and shares, and the track's inclusion in popular playlists.

A song could have high total streams but a lower popularity score if:

- Its streams peaked in the past and have slowed down recently (according to when the data was retrieved).
- Listener engagement is lower (song or artist frequently skipped).
- It's no longer being actively recommended or added to playlists.

```python
# Group by artist and sum the Spotify streams
artist_streams_2024 = data_2024.groupby('Artist')['Spotify Streams'].sum().reset_index()

# Sort the artists by total Spotify streams and get the top 10
top_2024_artists = artist_streams_2024.sort_values(by='Spotify Streams', ascending=False).head(10)

top_2024_artists
```

```python
top_2024_artists.plot(x='Artist', y='Spotify Streams', kind='bar', figsize=(10, 6), color='skyblue')
plt.title("Most played artists in 2024")
plt.xlabel("Artist")
plt.ylabel("Number of plays in billions")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

At the end, in order to select the most streamed artists in 2024 (who released music during this year). We see that Taylor Swift is in position number one even though when her songs were not in the top 5 of most streamed song or were not even rated as the most popular. However, she had a massive Tour during 2024 which might have influenced the streams of the songs.

Spotify Streams measure long-term success; Spotify Popularity captures current relevance.
Both metrics together offer a holistic view of a song's performance: streams for legacy and popularity for trends.

### Most streamed songs

```python
most_played_song = data_cleaned.loc[data_cleaned['Spotify Streams'].idxmax()]
most_played_song = most_played_song[['Track','Artist','Release Date', 'Spotify Streams']]
most_played_song
```

The most streamed song over time was 'Blinding Lights' by the Weeknd, with rough 4 billion streams, and being ranked in the 2nd position of artists over all time.

However, we noticed that some songs had been remasterized or had been relaunched in different versions, so in order to give a more comprehensive overview, we decided to sum up all the streams grouped by Track.

```python
song_artist_streams_by_year = data_cleaned.groupby([data_cleaned['Release Date'].dt.year, 'Track', 'Artist'])['Spotify Streams'].sum().reset_index()

song_artist_streams_by_year_sorted = song_artist_streams_by_year.sort_values(by=['Release Date', 'Spotify Streams'], ascending=[True, False])

top_song_per_year = song_artist_streams_by_year_sorted.drop_duplicates(subset='Release Date', keep='first')

top_song_per_year = top_song_per_year[['Release Date', 'Spotify Streams', 'Track', 'Artist']]

top_song_per_year
```

As Spotify was created in 2008, it is understandable that results prior to this date cannot be measured by grouping the years and selecting the most streamed artist. However, this piece of code showcases which songs are the most played to this day according to their release year.

```python
song_and_artist = data_cleaned[['Track', 'Artist','Release Date']]
song_streams_over_time = data_cleaned.groupby('Track')['Spotify Streams'].sum().reset_index()
merged = pd.merge(song_streams_over_time, song_and_artist.drop_duplicates(), on='Track',how='left')

most_streamed_songs = merged.sort_values(by='Spotify Streams', ascending=False)

most_streamed_songs.head(10)
```

This table showcases which are the most played songs and it is not surprise that some of these songs also appear in the previous chart that showcased the most streamed song per year.

```python
data_cleaned['Release Year'] = data_cleaned['Release Date'].dt.year

# Group by year and sum the Spotify streams
streams_per_year = data_cleaned.groupby('Release Year')['Spotify Streams'].sum().reset_index()

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(streams_per_year['Release Year'], streams_per_year['Spotify Streams'], marker='o', color='blue')
plt.title('Total Spotify Streams Per Year')
plt.xlabel('Year')
plt.ylabel('Total Spotify Streams in billions')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This graph showcases the growing trend in Spotify streams, however we see an inflexion point in 2020, probably because of corona. After that, the trend seem to be increasing. The dropdown in 2024 can be caused by the missing data in 2024.

# Key Business Insights & Decisions

### Most Streamed Songs & Artists
**Insight:** Identifies top-performing songs and artists.
**Business Decision:**

Music labels can invest more in high-performing artists through promotions, exclusive deals, and partnerships.
Emerging artists can learn from trending styles & strategies of top performers.

### Distribution of Song Popularity
**Insight:** Shows whether popularity is evenly distributed or if a few songs dominate.
**Business Decision:**

If only a few songs dominate, then Spotify/labels might diversify promotional strategies to support mid-tier artists.
Advertisers may target ads around popular songs, leveraging their visibility.
Adjust algorithms to balance new music discovery.

### Correlation Between Streams & Popularity (0.44 Correlation)
**Insight:** Popularity doesn’t fully determine streams — other factors matter (marketing, virality, social media, etc.).
**Business Decision:**

Labels should invest in multi-channel marketing (TikTok, YouTube, radio, playlists).
Emerging artists can still gain streams if they go viral—even if their “popularity” metric is low.
Spotify might rethink how they rank & recommend songs.

### Trend of Streams Over the Years
**Insight:** Tracks how streaming behavior has changed (e.g., peak years, growth rate).
**Business Decision:**

If streaming is increasing, platforms can monetize better (ads, subscriptions, exclusive releases).
If certain years had spikes, analyze why (e.g., did COVID-19 impact streaming habits?).
Predict future trends—should labels push more digital releases or explore other models?
